```r
library('gdata')
```

```r
mydata=read.csv('/Users/anandpopat/Desktop/2_grams.csv')
```

```r
mydata=mydata[2:3]
mydata=na.omit(mydata)
colnames(mydata)=c('files','time')
mydata
```

```r
library(ggplot2)
ggplot(data=mydata, aes(x=mydata$time, y=mydata$files, group=1)) +
  geom_line()+
  geom_point()
```
