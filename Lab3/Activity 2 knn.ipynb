```r
data=read.csv('german_credit_data1.csv')
```

```r
data1=na.omit(data)
```

```r
data2=data1[,c('Age','Saving.accounts','Checking.account','Credit.Risks')]
```

```r
data2$Saving.accounts=as.character(data2$Saving.accounts)
data2$Checking.account=as.character(data2$Checking.account)
```

Convert the categorical attributes checking and savings to numeric value.

```r
for (i in 1:nrow(data2)){
    if(data2[i,2]=='little'){
        data2[i,2]=sample(0:20,1)
    }
    if(data2[i,2]=='moderate'){
        data2[i,2]=sample(20:50,1)
    }
    if(data2[i,2]=='quite rich'){
        data2[i,2]=sample(50:80,1)
    }
    if(data2[i,2]=='rich'){
        data2[i,2]=sample(80:100,1)
    }
    if(data2[i,3]=='little'){
        data2[i,3]=sample(0:20,1)
    }
    if(data2[i,3]=='moderate'){
        data2[i,3]=sample(20:50,1)
    }
    if(data2[i,3]=='quite rich'){
        data2[i,3]=sample(50:80,1)
    }
    if(data2[i,3]=='rich'){
        data2[i,3]=sample(80:100,1)
    }
}
```

Make a category called income which is the sum of saving+checking

```r
data2$income=as.numeric(data2$Saving.accounts)+as.numeric(data2$Checking.account)
```

```r
data3=data2[,c('Age','income','Credit.Risks')]
data3
```

```r
#plot(data3$Age,data3$income)
```

Divide train and test set and perform knn for 25 times, keeping track of misclassification values to know which value of k is the best.

```r
n=nrow(data3)
rate=0.9
test_lab=n * (1 - rate)
training=sample(1:n,rate * n,replace=FALSE)
train=subset(data3[training, ], select = c(Age, income))
testing=setdiff(1:n, training)
test=subset(data3[testing, ], select = c(Age, income))
answer=data3$Credit.Risks[training]
real=data3$Credit.Risks[testing]
t=data.frame('k','Train=90%')
colnames(t)=c('k','Misclassification')
for (k in 1:25) {
    #print(k)
    predict=knn(train, test, answer, k)
    incorrect=sum(predict!=real) 
    misclassification=incorrect/test_lab
    t1=data.frame(as.character(k),as.character(misclassification))
    colnames(t1)=c('k','Misclassification')
    t=rbind(t,t1)
}
```
