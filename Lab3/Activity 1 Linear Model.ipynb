```r
library('gdata')
```

```r
data=read.xls('NHLTop100.xlsx')
```

```r
model_a <- lm (data$A~data$G)
summary(model_a)
plot(data$G, data$A, pch=20,col="red", xlab="Goals",
      ylab="Assists")
abline(model_a)
```

```r
y_b=head(data$A,1)
x_b=head(data$G,1)
model_b = lm(I(data$A-head(data$A,1)) ~ I(data$G-head(data$G,1)))
coefs<-coef(model_b)
m_b=coefs
c_b=y_b-m_b*x_b
model_b = lm(I(data$A-m_b[1]) ~ I(data$G))
summary(model_b)
plot(data$G, data$A, pch=20,col="red", xlab="Goals",
      ylab="Assists")
abline(model_b)
```

```r
patrick_G=285
patrick_A=462
patrick=data.frame('1','Patrick Kane','-','-','-','-','-',patrick_G,patrick_A,'-','-','-','-','-','-','-','-','-')
colnames(patrick)=colnames(data)
data1 <- rbind(data,patrick)
```

```r
y_c=patrick_A
x_c=patrick_G
model_c = lm(I(data1$A-y_c) ~ I(data1$G-x_c))
coefs<-coef(model_c)
m_c=coefs
c_c=y_c-m_c*x_c
model_c = lm(I(data1$A-m_c[1]) ~ I(data1$G))
summary(model_c)
plot(data1$G, data1$A, pch=20,col="red", xlab="Goals",
      ylab="Assists")
abline(model_c)
```

```r
model_d <- lm (data$A~data$G + 0)
summary(model_d)
plot(data$G, data$A, pch=20,col="red", xlab="Goals",
      ylab="Assists")
abline(model_d)
```

```r
y_e=head(data$A,1)
x_e=head(data$G,1)
model_e = lm(I(data$A-head(data$A,1)) ~ I(data$G-head(data$G,1))+0)
summary(model_e)
coefs<-coef(model_e)
m_e=coefs
c_e=y_e-m_e*x_e
plot(data$G, data$A, pch=20,col="red", xlab="Goals",
      ylab="Assists")
abline(c_e,m_e)
```

```r
y_f=patrick_A
x_f=patrick_G
model_f = lm(I(data1$A-y_f) ~ I(data1$G-x_f)+0)
summary(model_f)
coefs<-coef(model_f)
m_f=coefs
c_f=y_f-m_f*x_f
plot(data1$G, data1$A, pch=20,col="red", xlab="Goals",
      ylab="Assists")
abline(c_f,m_f)
```
